### Encode categorical values

```python
df2 = df1.copy()
```

```python
make_le = LabelEncoder()
drive_wheels_le = LabelEncoder()
engine_location_le = LabelEncoder()
num_of_cylinders_le = LabelEncoder()
horsepower_binned_le = LabelEncoder()
```

```python
df2['make'] = make_le.fit_transform(df1['make'])
df2['drive_wheels'] = drive_wheels_le.fit_transform(df1['drive_wheels'])
df2['engine_location'] = engine_location_le.fit_transform(df1['engine_location'])
df2['num_of_cylinders'] = num_of_cylinders_le.fit_transform(df1['num_of_cylinders'])
df2['horsepower_binned'] = horsepower_binned_le.fit_transform(df1['horsepower_binned'])
```

### 2. Using other following methods
* Balanced Bagging Classifier
* Logistic Regression

```python
x = mdf.drop(['price'] , axis = 1).values
y = mdf['price'].values
```

### Split Data

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size= 0.25 , random_state=42)
[x_train.shape, y_train.shape]
```

### Data Scaling using Robust Scaler

```python
ro_scaler = RobustScaler()
x_train = ro_scaler.fit_transform(x_train)
x_test = ro_scaler.fit_transform(x_test)
[x_train.shape, x_test.shape]
```

### Balanced Bagging Classifier

```python
bbc_classifier = BalancedBaggingClassifier(base_estimator=DecisionTreeClassifier(), sampling_strategy='not majority', replacement=False, random_state=42)
```

```python
bbc_classifier.fit(x_train, y_train)
```

```python
bbc_classifier.score(x_test , y_test)
```

```python
y_pred = bbc_classifier.predict(x_test)
data = pd.DataFrame({"Y_test" : y_test , "y_pred1": y_pred})
data.head(10)
```

```python
plt.figure(figsize=(10,8))
plt.plot(data[:100])
plt.legend(["Actual" , "Predicted"])
```

### Logistic Regression

```python
lr = LogisticRegression(C = 100)
```

```python
lr.fit(x_train , y_train)
```

```python
lr.score(x_train , y_train)
```

```python
lr.score(x_test , y_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred1 = lr.predict(x_test)
data = pd.DataFrame({"Y_test" : y_test , "y_pred1": y_pred1})
data.head(10)
```

```python
plt.figure(figsize=(10,8))
plt.plot(data[:100])
plt.legend(["Actual" , "Predicted"])
```
