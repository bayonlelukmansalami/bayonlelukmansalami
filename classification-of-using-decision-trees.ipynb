### Train Test Split

Out of Sample Accuracy is the percentage of correct predictions that the model makes on data that the model has NOT been trained on. Doing a train and test on the same dataset will most likely have low out-of-sample accuracy, due to the likelihood of our model overfitting.

It is important that our models have a high, out-of-sample accuracy, because the purpose of any model, of course, is to make correct predictions on unknown data. So how can we improve out-of-sample accuracy? One way is to use an evaluation approach called Train/Test Split.
Train/Test Split involves splitting the dataset into training and testing sets respectively, which are mutually exclusive. After which, you train with the training set and test with the testing set.

This will provide a more accurate evaluation on out-of-sample accuracy because the testing dataset is not part of the dataset that has been used to train the model. It is more realistic for the real world problems.


```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

<h3> Setting Up The Decision Trees</h3>

### Modeling
We will first create an instance of the DecisionTreeClassifier called drugTree.
Inside of the classifier, specify criterion="entropy" so we can see the information gain of each node.

```python
deathTree = DecisionTreeClassifier(criterion="entropy", max_depth = 6)
deathTree # it shows the default parameters
```

```python
deathTree.fit(X_train,y_train)
```

### Prediction
Let's make some predictions on the testing dataset and store it into a variable called predTree.

```python
predTree = deathTree.predict(X_test)
```

You can print out predTree and y_testset if you want to visually compare the predictions to the actual values.

```python
print (predTree [0:10])
print (y_test [0:10])
```

#### Confusion Matrix

```python
confusion_matrix = metrics.confusion_matrix(y_test, predTree)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.show()
```

```python
Accuracy = metrics.accuracy_score(y_test, predTree)
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

```python
Precision = metrics.precision_score(y_test, predTree)
```

```python
print("DecisionTrees's Precision: ", metrics.precision_score(y_test, predTree))
```

```python
Sensitivity = metrics.recall_score(y_test, predTree)
```

```python
print("DecisionTree's Sensitivity_recall: ",metrics.recall_score(y_test, predTree) )
```

```python
Specificity = metrics.recall_score(y_test, predTree, pos_label=0)
```

```python
print("DecisionTree's Specificity: ",metrics.recall_score(y_test, predTree, pos_label = 0))
```

```python
F1_score = metrics.f1_score(y_test, predTree)
print("DecisionTree's F1 Score: ",metrics.f1_score(y_test, predTree))
```

**Accuracy classification score** computes subset accuracy: the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true.

In multilabel classification, the function returns the subset accuracy. If the entire set of predicted labels for a sample strictly matches with the true set of labels, then the subset accuracy is 1.0; otherwise it is 0.0.
